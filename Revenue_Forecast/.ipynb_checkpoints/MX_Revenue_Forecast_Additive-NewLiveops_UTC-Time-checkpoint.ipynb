#### Functions

```python
# Get Confidence Interval
def getCI(x,y,danger):  
    n = len(x)
    sum_errs = np.sum((x - y)**2)
    stdev = np.sqrt(1/(len(x)-2) * sum_errs)
    Conf_Interval = stdev * stats.t.ppf((1 + danger) / 2, n - 1)
    return Conf_Interval

def fourierExtrapolation(x, n_predict, harmonics, keep_trend=True):
    n = x.size
    n_harm = harmonics                     # number of harmonics in model
    t = np.arange(0, n)
    p = np.polyfit(t, x, 1)         # find linear trend in x
    x_notrend = x - p[0] * t        # detrended x
    x_freqdom = fft.fft(x_notrend)  # detrended x in frequency domain
    f = fft.fftfreq(n)              # frequencies
    indexes = list(range(n))            
    #indexes.sort(key = lambda i: np.absolute(f[i]))      # sort indexes by frequencies, lower -> higher
    indexes.sort(key=lambda i: np.absolute(x_freqdom[i])) # sort indexes by amplitude, lower -> higher
    indexes.reverse()    
    t = np.arange(0, n + n_predict)
    restored_sig = np.zeros(t.size)
    for i in indexes[:1 + n_harm * 2]:
        ampli = np.absolute(x_freqdom[i]) / n   # amplitude
        phase = np.angle(x_freqdom[i])          # phase
        restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase)
    if keep_trend:
        return restored_sig + p[0] * t
    else:
        return restored_sig
```

```python
# Predicts Train and Test
rf = RandomForestRegressor(n_estimators=100, criterion='mse', oob_score = True, bootstrap = True, random_state = 0)
rf.fit(r_train[:len(noise)], noise)
predict_RF_train = rf.predict(r_train)
rmse = mean_squared_error(noise, predict_RF_train[:len(noise)])/(noise.mean()**int(noise.mean()!=0))
R2_Score_RF, OOB_RF, RMSE_RF= rf.score(r_train[:len(noise)], noise), rf.oob_score_, rmse
print('R^2 Training Score: {:.3f} \nOOB Score: {:.3f} \nRMSE: {:.3f}'.format(R2_Score_RF,OOB_RF,RMSE_RF))
```
